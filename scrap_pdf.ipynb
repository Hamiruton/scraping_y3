```python
import fitz, PyPDF2, time
```

```python
def extract_pdf(pdf_file):
  with open(pdf_file, 'rb') as file:
    pdf_reader = PyPDF2.PdfReader(file)

    page = pdf_reader.pages[6]
    text = page.extract_text()
  
  return text
```

```python
pdf_file_path = "journal_hier.pdf"
pdf_text = extract_pdf(pdf_file_path)
```

```python
pdf_doc = fitz.open(pdf_file_path)
page = pdf_doc[6]

width, height = page.mediabox_size

x1 = 0
y1 = height / 2
x2 = width
y2 = height

pixmap = page.get_pixmap(matrix=fitz.Matrix(2,2), clip=(x1,y1,x2,y2))
screenshot_filename = "Capture_page_5.png"
pixmap.save(screenshot_filename)
texte = page.get_text('text')
print(texte)
print(f'Capture enregistrée : {screenshot_filename}')
pdf_doc.close()
```

```python
def extract_acc_text(pdf_file_path, text_to_find):
  with fitz.open(pdf_file_path) as pdf_doc:
    page = pdf_doc[6]
    texte = page.get_text('text')
    with open('new_fichier_encoded.txt', 'w') as file:
      texte = str(texte).lower()
      new_texte = texte.replace("’", "'")
      file.write(new_texte)
    find_word = new_texte.find(text_to_find)
    print(find_word)
```

```python
extract_acc_text(pdf_file_path, "appel d'offres")
```

```python
# 2nd function : screen the pages which contain the keyword "appel d'offres"
def screen_page(page, num_page):
  pixmap = page.get_pixmap(matrix=fitz.Matrix(2,2))
  screenshot_filename = f"journal_screenshot_{num_page}.png"
  pixmap.save(screenshot_filename)
  print(f'Capture enregistrée : {screenshot_filename}')
    
```

```python
# 1st function : browse the pages of newspaper for the keyword "appel d'offres"
def browse_pages(pdf_file, keyword = "appel d'offres"):
  with fitz.open(pdf_file) as pdf_doc:
    for page_num in range(pdf_doc.page_count):
      page = pdf_doc[page_num]

      text = page.get_text('text')
      formatted_text = str(text).lower()
      formatted_text = formatted_text.replace("’", "'")

      is_keyword = formatted_text.find(keyword)
      if is_keyword != -1 :
        screen_page(page, page_num)
```

```python
pdf_file_path = "journaux/echos_2.pdf"
browse_pages(pdf_file_path)
```
